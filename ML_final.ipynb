```python
import pandas as pd
df_sales=pd.read_csv("sales_train_evaluation.csv")
df_sales.head()
```

```python
df_cal=pd.read_csv("calendar.csv")
df_cal.head()
```

```python
df_cal[pd.notna(df_cal['event_name_1'])]
```

```python
df_cal['event_type_1'].unique()
```

```python
df_cal = df_cal.rename(columns={'d': 'day'})
df_cal.head()
```

```python
df_cal['event_type_1'].fillna('None', inplace=True)
df_cal['event_type_2'].fillna('None', inplace=True)
df_cal.head()
```

```python
df_cal1=df_cal[df_cal['year']>2013]
a=df_cal1['day'].unique()
col=[]
for i in a:
    col.append(i)
id_col=['id','item_id','dept_id','cat_id','store_id','state_id']
columns=id_col+col
df_sales=df_sales[(df_sales['state_id']=='CA') & (df_sales['cat_id']=='HOBBIES')][columns[:879]]
df_sales.shape
```

```python
df_sales.head(10)
```

```python
df_sales_t=df_sales.T
df_sales_t=df_sales_t[6:]
df_sales_t=df_sales_t.reset_index()
df_sales_t = df_sales_t.rename(columns={'index': 'day'})

df_sales_t.head()
```

```python
df_sales_t.tail()
```

```python
for column in df_sales_t.columns[1:]:  # Skip the first column which contains the 'day'
    # Calculate the 75th percentile for the current product
    percentile_75 = df_sales_t[column].quantile(0.75)

    # Cap values at the 75th percentile
    df_sales_t[column] = df_sales_t[column].clip(upper=percentile_75)

    # Floor values at 0
    df_sales_t[column] = df_sales_t[column].clip(lower=0)
```

```python
df_table = pd.merge(df_sales_t, df_cal[['wday', 'month','snap_CA', 'day','event_type_1','event_type_2']], on='day', how='inner')
df_table=df_table.drop('day', axis=1)
df_table.head()
```

```python
df_table.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# Separate the numerical and categorical columns
numerical_cols = df_table.columns[:-5]  # All columns except the last 5
categorical_cols = ['wday', 'month', 'snap_CA','event_type_1','event_type_2']

# Normalize the sales data
scaler = MinMaxScaler(feature_range=(0, 1))
df_table[numerical_cols] = scaler.fit_transform(df_table[numerical_cols])

# Cyclical encoding for 'wday' and 'month'
df_table['wday_sin'] = np.sin(df_table['wday'] * (2. * np.pi / 7))
df_table['wday_cos'] = np.cos(df_table['wday'] * (2. * np.pi / 7))
df_table['month_sin'] = np.sin((df_table['month'] - 1) * (2. * np.pi / 12))
df_table['month_cos'] = np.cos((df_table['month'] - 1) * (2. * np.pi / 12))

# Drop the original 'wday' and 'month' columns
df_table.drop(['wday', 'month'], axis=1, inplace=True)
```

```python
df_table.shape
```

```python
df_table.head()
```

```python
encoded_df = pd.get_dummies(df_table[['event_type_1', 'event_type_2']])

# Concatenate the encoded columns with the original DataFrame
df_table = pd.concat([df_table, encoded_df], axis=1)

# Drop the original 'event_type_1' and 'event_type_2' columns
df_table.drop(['event_type_1', 'event_type_2'], axis=1, inplace=True)
```

```python
df_table.head()
```

```python
df_table.shape
```

```python
import numpy as np

# Total number of days used for training
total_days = 845

# Length of the sequence (28 days)
sequence_length = 28

# Prepare the lists for x_train and y_train
x_train = []
y_train = []

# Loop to create sequences for training
for i in range(total_days - sequence_length):
    # Include both sales data and categorical features for x_train
    x_train.append(df_table.iloc[i:i + sequence_length].values)
    # y_train is still the sales data of the next day
    y_train.append(df_table.iloc[i + sequence_length, :-13].values)

# Convert lists to numpy arrays
x_train = np.array(x_train)
y_train = np.array(y_train)

# Checking the shapes
x_train.shape, y_train.shape
```

```python
x_train_final = x_train[:-28]  # Exclude the last 28 sequences for validation
y_train_final = y_train[:-28]  # Exclude the last 28 targets for validation

x_val = x_train[-28:]  # Last 28 sequences for validation
y_val = y_train[-28:]  # Last 28 targets for validation
```

```python
y_train_final.shape
```

```python
x_train_final.shape
```

Modelling stage


```python
# Model 1 with results
# 2 Bidirectional LSTM layers each with 100 neurons
# Adam optimizer with learning rate of 0.01
# Only 10 epochs of batch size 32


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional

# Define a more complex LSTM model
model = Sequential()

# Adding the first Bidirectional LSTM layer
model.add(Bidirectional(LSTM(units=100, return_sequences=True), input_shape=(x_train_final.shape[1], x_train_final.shape[2])))
model.add(Dropout(0.3))

# Adding a second Bidirectional LSTM layer
model.add(Bidirectional(LSTM(units=100, return_sequences=False)))
model.add(Dropout(0.3))

# Adding a Dense layer
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.3))

# Adding the output layer
model.add(Dense(units=y_train_final.shape[1]))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(x_train_final, y_train_final, epochs=10, batch_size=32, validation_data=(x_val, y_val), verbose=1)
```

```python
predictions = []
current_sequence = x_train[-1].reshape(1, 28, 2273)

for day in range(28):
    # Predict the next day's sales
    predicted_sales = model.predict(current_sequence)

    # Reshape predicted_sales to be 3-dimensional
    predicted_sales = predicted_sales.reshape(1, 1, -1)

    # Fetch the corresponding categorical features for the next day
    next_day_categorical = df_table.iloc[844 + day, -13:].values.reshape(1, 1, 13)

    # Combine predicted sales with the categorical features
    predicted_sales_with_cat = np.concatenate((predicted_sales, next_day_categorical), axis=2)

    # Append the prediction to the predictions list (excluding categorical features)
    predictions.append(predicted_sales)

    # Update the sequence for the next prediction
    next_sequence = current_sequence[:, 1:, :]  # Remove the first day
    next_sequence = np.concatenate((next_sequence, predicted_sales_with_cat), axis=1)  # Append the prediction with categorical features

    current_sequence = next_sequence

# Convert predictions to a numpy array
predictions = np.array(predictions).reshape(28, 2260)  # Only sales data
```
